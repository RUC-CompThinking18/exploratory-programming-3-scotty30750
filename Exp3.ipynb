```python
import re
source = open('book.txt')
fish = source.read()

def read_book(book):
    if type(book) != str:
        raise TypeError("Input is not a string")
    match =re.findall(r"[\w]*at\b", book)
    matches = []
    for m in match:
        # need to precede the Regex string with r
        # this tell re that the string is a regex
        if len(m) > 3:
            matches.append(m)
    return matches
```

```python
read_book(fish)
```
